```python
## A Plataforma que possui mais vendas Globais são Publicadas pela Nintendo?
```

```python
# Filtrar apenas os jogos publicados pela Nintendo
nintendo_games = vg[vg['Publisher'] == 'Nintendo']
```

```python
# Agrupar por plataforma e somar as vendas globais
platform_sales = nintendo_games.groupby('Platform')['Global_Sales'].sum().reset_index()
```

```python
# Encontrar a plataforma com mais vendas globais
max_sales_platform = platform_sales.loc[platform_sales['Global_Sales'].idxmax()]
```

```python
print(f"A plataforma com mais vendas globais publicada pela Nintendo é: {max_sales_platform['Platform']} com {max_sales_platform['Global_Sales']} milhões de unidades vendidas.")
```

```python
## O gênero de jogo com mais vendas globais é "Sports"? 
```

```python
# Agrupar por gênero e somar as vendas globais
genre_sales = vg.groupby('Genre')['Global_Sales'].sum().reset_index()
```

```python
genre_sales
```

```python
# Encontrar o gênero com mais vendas globais
max_sales_genre = genre_sales.loc[genre_sales['Global_Sales'].idxmax()]
```

```python
max_sales_genre
```

```python
# Verificar se o gênero com mais vendas globais é "Sports"
is_sports = max_sales_genre['Genre'] == 'Sports'

if is_sports:
    print(f"O gênero com mais vendas globais é 'Sports' com {max_sales_genre['Global_Sales']:0.2f} milhões de unidades vendidas.")
else:
    print(f"O gênero com mais vendas globais não é 'Sports', é '{max_sales_genre['Genre']}' com {max_sales_genre['Global_Sales']:0.2f} milhões de unidades vendidas.")
```
