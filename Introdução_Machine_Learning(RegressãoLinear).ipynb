```python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# URL bruto do arquivo CSV no GitHub
url = 'https://raw.githubusercontent.com/gihAzevedo/SantanderCoders/main/USA_Housing.csv'

# Carregar o dataset em um DataFrame
df = pd.read_csv(url)
```

```python
df.head()
```

```python
casas_numericas = df.select_dtypes(include=['float64', 'int64'])
```

```python
casas_numericas.corr()
```

```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Area Population']] # caracteristicas
y = df['Price'] # alvo
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x_treino, y_treino)
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
y_previsto
```

```python
df['Price'].mean()
```

```python
mean_absolute_error(y_previsto, y_teste)
```
